- Con la intención de hacer un manejo más sencillo de los datos se van a modificar con la idea de crear cuatro columnas con los datos originales (State - date - Months - Precipitation).
- También se crea una quinta columna calculando las diferencias de las Precipitaciones entre cada dia para cada año, mes y estado de EEUU.

```python
#Se crean columnas rellenadas con los meses del año

df['January_1'] = 'January'
df['February_1'] = 'February'
df['March_1'] = 'March'
df['April_1'] = 'April'
df['May_1'] = 'May'
df['June_1'] = 'June'
df['July_1'] = 'July'
df['August_1'] = 'August'
df['September_1'] = 'September'
df['October_1'] = 'October'
df['November_1'] = 'November'
df['December_1'] = 'December'
```

```python
#se crean variables uniendo cada mes con sus valores correspondientes de pp

df_1 = df[['State', 'date', 'January_1', 'January']]
df_1.rename(columns = {'January_1':'Months', 'January': 'Precipitation'}, inplace = 1)

df_2 = df[['State', 'date', 'February_1', 'February']]
df_2.rename(columns = {'February_1':'Months', 'February': 'Precipitation'}, inplace = 1)

df_3 = df[['State', 'date', 'March_1', 'March']]
df_3.rename(columns = {'March_1':'Months', 'March': 'Precipitation'}, inplace = 1)

df_4 = df[['State', 'date', 'April_1', 'April']]
df_4.rename(columns = {'April_1':'Months', 'April': 'Precipitation'}, inplace = 1)

df_5 = df[['State', 'date', 'May_1', 'May']]
df_5.rename(columns = {'May_1':'Months', 'May': 'Precipitation'}, inplace = 1)

df_6 = df[['State', 'date', 'June_1', 'June']]
df_6.rename(columns = {'June_1':'Months', 'June': 'Precipitation'}, inplace = 1)

df_7 = df[['State', 'date', 'July_1', 'July']]
df_7.rename(columns = {'July_1':'Months', 'July': 'Precipitation'}, inplace = 1)

df_8 = df[['State', 'date', 'August_1', 'August']]
df_8.rename(columns = {'August_1':'Months', 'August': 'Precipitation'}, inplace = 1)

df_9 = df[['State', 'date', 'September_1', 'September']]
df_9.rename(columns = {'September_1':'Months', 'September': 'Precipitation'}, inplace = 1)

df_10 = df[['State', 'date', 'October_1', 'October']]
df_10.rename(columns = {'October_1':'Months', 'October': 'Precipitation'}, inplace = 1)

df_11 = df[['State', 'date', 'November_1', 'November']]
df_11.rename(columns = {'November_1':'Months', 'November': 'Precipitation'}, inplace = 1)

df_12 = df[['State', 'date', 'December_1', 'December']]
df_12.rename(columns = {'December_1':'Months', 'December': 'Precipitation'}, inplace = 1)
```

```python
# concateando de todas las variables

df_concat = pd.concat([df_1, df_2, df_3, df_4, df_5, df_6, df_7, df_8, df_9, df_10, df_11, df_12])
```

```python
#Paso de pulgadas de lluvia (medida estadounidense) a milimetros.

df_concat['Precipitation'] = (df_concat['Precipitation']* 25.4)
```
